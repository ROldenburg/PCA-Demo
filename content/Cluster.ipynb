# Clusternanalyse #

Dieses Notebook zeigt die Durchführung einer k-Means Clusteranalyse. Die Datenvektoren werden aus der Bibliothek numpy verwendet. Insbesondere benötigt man folgende numpy-Befehle:

np.asarray(L) verwandelt eine in Listenstruktur dargestellte Matrix in eine numpy-Matrix

np.dot(u,v) Skalarprodukt zweier Vektoren

np.linalg.norm(v) Norm eines Vektors

```python
import numpy as np
BeispielDaten=[np.array([9,2,3]),np.array([9,9,9]),np.array([7,0,5]),
               np.array([8,4,4]),np.array([9,3,2]),np.array([7,9,8]),
               np.array([1,1,1]),np.array([2,2,3]),np.array([-1,2,2])]
# Berechnung des Mittelpunktes einer Lsite von Vektoren
def MP(daten:list,fallback=np.array([0,0,0]))->list:
    if len(daten)==0: # Leere Daten haben kein Zentrum, dann fallback
        return fallback
    return sum(daten)/len(daten)
MP(BeispielDaten)
```

Jetzt kommt der eigentliche Algorithmus

```python
import random
def kCluster(daten:list,k:int):
    n=len(daten)
    clusterVon=[i%k for i in range(n)] # Willkürliche Startzuordnung
    def PktvClus(j:int)->list: # j ist Cluster-Nummer
        return [daten[i] for i in range(n) if clusterVon[i]==j] 
    geändert=True     
    while geändert:
        geändert=False
        zentren=[MP(PktvClus(j),fallback=random.choice(daten)) for j in range(k)]         
        for i in range(n): # Punkte den Zentren zuordnen
            punkt=daten[i]             
            optimal=0 # Nummer des besten Zentrums 
            for c in range(1,k):
                if (np.linalg.norm(zentren[c]-punkt)<np.linalg.norm(zentren[optimal]-punkt) ):
                    optimal=c        
            if not(clusterVon[i]==optimal):
                geändert=True             
                clusterVon[i]=optimal     
    return [PktvClus(j) for j in range(k)] 
kCluster(BeispielDaten,3)
```
